# Manipulating time series data by Data Camp
## Part 1 : Working with Time Series in Pandas

```python
import pandas as pd  # assumed imported going forward
from datetime import datetime  # To manually create dates
```

#### Timestamp object has many attributes to store time-speci cinformation

```python
time_stamp = pd.Timestamp(datetime(2017, 1, 1))

# Understands dates as strings
print(f"{pd.Timestamp('2017-01-01') == time_stamp }")
print(f"{time_stamp.year}")
print(f"{time_stamp.day_name()}")
```

#### Period object has freqattribute to store frequencyinfo

```python
period = pd.Period('2017-01')
print(period.asfreq('D'))

# Convert pd.Period()to pd.Timestamp() andback
print(period.to_timestamp().to_period('M'))

# Frequency info enables basic date arithmetic
print(period.asfreq('D') + 2)
print(period + 2)
```

#### Sequences of dates & times

```python
index = pd.date_range(start='2017-1-1', periods=12, freq='M')

# Convert to period
index.to_period()# Create the range of dates here
```

Your first time series

```python
# Create the range of dates here
seven_days = pd.date_range(start='2017-1-1', periods=7, freq='D')

# Iterate over the dates and print the number and name of the weekday
for day in seven_days:
    print(day.dayofweek, day.day_name())
```

### Create a time series of air quality data

```python
import matplotlib.pyplot as plt
data = pd.read_csv('./air_quality_data/nyc.csv')

# Inspect data
print(data.info())

# Convert the date column to datetime64
data.date = pd.to_datetime(data.date)

# Set date column as index
data.set_index("date", inplace=True)

# Inspect data 
print(data.info())
```

### Compare annual stock price trends

```python
yahoo = pd.read_csv("./stock_data/yahoo.csv")

# Convert the date column to datetime64
yahoo.date = pd.to_datetime(yahoo.date)

# Set date column as index
yahoo.set_index("date", inplace=True)

# Create dataframe prices here
prices = pd.DataFrame()

# Select data for each year and concatenate with prices here 
for year in ["2013", "2014", "2015"]:
    price_per_year = yahoo.loc[year, ["price"]].reset_index(drop=True)
    price_per_year.rename(columns={"price": year}, inplace=True)
    prices = pd.concat([prices, price_per_year], axis=1)

# Plot prices
prices.head()

# Plot data
prices.plot(subplots=True)
plt.rcParams['figure.figsize'] = [20, 7]
plt.tight_layout(pad=2)
plt.show()
```

### Set and change time series frequency

```python
co = pd.read_csv("./air_quality_data/co_cities.csv")

# Convert the date column to datetime64
co.date = pd.to_datetime(co.date)

# Set date column as index
co.set_index("date", inplace=True)

# Set the frequency to calendar daily
co = co.asfreq("D")

# Plot the data
co.plot(subplots=True)
plt.tight_layout(pad=2)
plt.show()

# Set frequency to monthly
co = co.asfreq("M")

# Plot the data

co.plot(subplots=True)
plt.tight_layout(pad=2)
plt.show()
```

### Lags, changes, and returns for stock price series

```python
google = pd.read_csv("./stock_data/google.csv", parse_dates=["Date"], index_col=["Date"])

# Set data frequency to business daily
google = google.asfreq("B")

google['shifted'] = google.Close.shift(periods=1)
google['lagged'] = google.Close.shift(periods=-1)

# Calculate one-period percent change Xt / Xt-1
google['change'] = google.Close.div(google.shifted)

# Calculate one-period percent change
google['return'] = google.change.sub(1).mul(100)

# Calculate three-period percent change
google['return_3d'] = google.Close.pct_change(periods=3).mul(100)

# Calculate diff in value for two adjacent periods
google['diff'] = google.Close.diff()

# Percent change for two adjacent periods Xt / Xt-1
google['pct_change'] = google.Close.pct_change().mul(100)
```

### Shifting stock prices across time

```python
google['shifted'] = google.Close.shift(periods=90) 
google['lagged'] = google.Close.shift(periods=-90)


google[["shifted", "lagged", "Close"]].plot(subplots=True)
plt.tight_layout(pad=2)
plt.show()
```

### Calculating stock price changes

```python
yahoo = pd.read_csv("./stock_data/yahoo.csv", parse_dates=["date"], index_col=["date"])
# Set data frequency to business daily
yahoo = yahoo.asfreq("B")
```

```python
# Created shifted_30 here
yahoo['shifted_30'] = yahoo.price.shift(periods=30)

# Subtract shifted_30 from price
yahoo['change_30'] =  yahoo.price.sub(yahoo.shifted_30)

# Get the 30-day price difference
yahoo['diff_30'] = yahoo.price.diff(periods=30)

# Inspect the last five rows of price
print(yahoo.tail(5))

# Show the value_counts of the difference between change_30 and diff_30
print(yahoo['change_30'].sub(yahoo['diff_30']).value_counts())
```

### Plotting multi-period returns

```python
google = pd.read_csv("./stock_data/google.csv", parse_dates=["Date"], index_col=["Date"])

# Set data frequency to business daily
google = google.asfreq("B")
```

```python
# Create daily_return
google['daily_return'] = google.Close.pct_change(1).mul(100)

# Create monthly_return
google['monthly_return'] = google.Close.pct_change(30).mul(100)

# Create annual_return
google['annual_return'] = google.Close.pct_change(360).mul(100)

# Plot the result
google.plot(subplots=True)
plt.tight_layout(pad=2)
plt.show()
```
